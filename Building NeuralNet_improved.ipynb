```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# Network and Training
EPOCHS = 200
BATCH_SIZE = 128
VERBOSE= 1
NB_CLASSES=10
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2
DROPOUT = 0.3
```

```python
mnist = keras.datasets.mnist
(X_train, Y_train,), (X_test, Y_test) = mnist.load_data()
```

```python
RESHAPED = 28*28
```

```python
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
# Normalize inputs within [0, 1].
X_train, X_test = X_train / 255.0, X_test / 255.0
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# One-hot representations for labels.
Y_train = tf.keras.utils.to_categorical(Y_train, NB_CLASSES)
Y_test = tf.keras.utils.to_categorical(Y_test, NB_CLASSES)
```

```python
# Building the Model
model = tf.keras.models.Sequential(
    [
        keras.layers.Dense(N_HIDDEN, activation='relu', name="dense_layer"),
        keras.layers.Dropout(DROPOUT),
        keras.layers.Dense(N_HIDDEN, activation='relu', name="dense_layer1"),
        keras.layers.Dropout(DROPOUT),
        keras.layers.Dense(10, activation='softmax', name="dense_layer2")
    ]
)
```

```python
model.build(input_shape=X_train.shape)
```

```python
model.summary()
```

```python
# compile the model
model.compile(
    optimizer='RMSProp',
    loss="CategoricalCrossentropy",
    metrics=["accuracy"]
)
```

```python
# training the model
model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
```

```python
# Evaluating the model
test_loss, test_acc = model.evaluate(X_test, Y_test)
print("Test accuracy: ", test_acc)
```
